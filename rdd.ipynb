```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
G.add_edges_from([('X', 'U'), ('U', 'X'), ('X', 'D'), ('U', 'D'), ('X', 'Y'), ('U', 'Y'), ('D', 'Y')])
pos = {'X': (0, 1), 'U': (1, 2), 'D': (0, 0), 'Y': (1, 0)}
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, arrows=True)
plt.title('Data Generating Graph (A)')
plt.axis('off')
plt.show()
G2 = nx.DiGraph()
G2.add_edges_from([('X', 'c0'), ('c0', 'U'), ('U', 'D'), ('U', 'Y'), ('D', 'Y')])
pos2 = {'X': (0, 1), 'c0': (0.5, 1.5), 'U': (1, 2), 'D': (0, 0), 'Y': (1, 0)}
plt.figure(figsize=(8, 6))
nx.draw(G2, pos2, with_labels=True, node_color='lightgreen', node_size=1500, arrows=True)
plt.title('Limiting Graph (B)')
plt.axis('off')
plt.show()
```
